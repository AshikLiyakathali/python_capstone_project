```python
import os
import time
import pandas_gbq
import pandas as pd

from datetime import datetime
from google.cloud import bigquery
```

```python
def main(project, query_str):
    """
    This function is the entry point function to fetch the individual files and merge them into a single dataframe.
    """
    client = bigquery.Client(project)
 
    dataframe = (
        client.query(query_str)
        .result()
        .to_dataframe(
        create_bqstorage_client=True,
        )
    )
    return dataframe
```

```python
project = 'gcp-ent-capstone-dev'
query_str = """
SELECT * 
FROM `gcp-ent-capstone-dev.TestTeam4.d_date_workdays`
"""

df = main(project, query_str)
df
```

```python
df.to_gbq('TestTeam4.d_date_workdays_test2', 'gcp-ent-capstone-dev', 'us-east1', if_exists = 'replace')
```
